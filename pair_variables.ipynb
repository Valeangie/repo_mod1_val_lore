```python
#1.4 Crea dos variables numéricas de tipo float (decimal) con dos decimales, llámalas de la siguiente forma:
altura1= 1.75
altura2= 1.60
#3.1 Cuál es la suma de las dos alturas
altura1 + altura2
```

```python
#1.5 Cuál es la diferencia entre las alturas
diferencia_altura = altura1 - altura2 
print(diferencia_altura)
```

```python
#2 Escribe un programa usando la función input() que le pregunte al usuario su nombre y apellidos. Una vez que tengamos el nombre del usuario:
#Imprime su nombre y apellidos todo en mayúsculas
nombre = input("cual es tu nombre y apellido?").upper()
print(nombre)
```

```python
#2.1 Imprime su nombre y apellidos todo en minúsculas
nombre = input("cual es tu nombre y apellido?").lower()
print(nombre)
```

```python
#2.2 Imprime su nombre y apellidos con la primera letra del nombre y los apellidos en mayúscula y el resto en minúscula.
nombre = input("cual es tu nombre y apellido?").title()
print(nombre)
```

```python
#2.3 Imprime solo la primera letra del nombre en mayúscula.
nombre = input("cual es tu nombre y apellido?").capitalize()
print(nombre)
```
